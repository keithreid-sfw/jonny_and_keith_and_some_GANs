```python
# View
def get_mode():
    mode = str(input("""Please enter a letter...
                            Would you like to generate a _random Bowie or _Load one?
                            r/L: """))
    return mode

def show_this_bowie(this_bowie):
    bowie_length = 16 # special variable
    for i in range(bowie_length):
        for j in range(bowie_length):
            if this_bowie[i][j] == 1:
                plt.scatter(j,16-i)
    plt.show()
```

```python
                
# Controller

if __name__ == "__main__":
    print("Bowie")
    mode                = get_mode()
    reference_bowies    = get_reference_bowies()
    
    if mode == "r":
        print("\nrandom bowie...")
        this_bowie = build_random_bowie()
    else:
        print("\nloading bowie...")
        which_bowie = int(input("""Please enter a number...
                                      load referance bowie 1,2,3,4 or your own in whci case 9?
                                      1 (default) , 2,3,4,9: """))
        this_bowie = reference_bowies[which_bowie]
    
    show_this_bowie(this_bowie)
    similarity_scores = []
    for reference_bowie in reference_bowies:
        similarity_score = compare_bowies(this_bowie, reference_bowie)
        similarity_scores.append(similarity_score)
    best_fit_index = np.argmax(similarity_scores)
    names = ["Heroes","Aladdin Sane","Low","Hunky Dory"]
    print("Best fit was:\t", names[best_fit_index])
    best_bowie = reference_bowies[best_fit_index]
    show_this_bowie(best_bowie)
                            
        
                             
        
    
```
